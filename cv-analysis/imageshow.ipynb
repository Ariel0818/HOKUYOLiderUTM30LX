```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
data_matrix = np.load('DepthCamera_Data_Tanya/20 FEB 2023/1_2DepthData.npy')
```

```python
x, y = np.meshgrid(range(data_matrix.shape[1]), range(data_matrix.shape[0]))
# 使用矩阵中的值作为z坐标
z = data_matrix

# 方块的大小
dx = np.full((420, 560), 0.05)
dy = np.full((420, 560), 0.05)
dz = np.full((420, 560), 0.05)
```

## 使用cube来投影到三维空间，耗时非常长，并且看不出啥来

```python
# 将数组扁平化
x = x.ravel()
y = y.ravel()
z = z.ravel()
dx = dx.ravel()
dy = dy.ravel()
dz = dz.ravel()
```

```python
# large calculating time
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(x, y, z, dx, dy, dz)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

---

## 使用sactter点云函数模拟，也比较一般

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='plasma', s=4)  # s 调整点的大小 由z值的大小调整颜色
plt.show()
```

## 直接绘制表面图--3D

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制表面
ax.plot_surface(x, y, z, cmap='viridis')

# 设置坐标轴标签
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 显示图形
plt.show()
```

## 直接绘制表面图--2D

```python
# 使用 imshow 展示数据
plt.imshow(z, cmap='viridis', origin='lower', extent=[0, z.shape[1], 0, z.shape[0]])
plt.colorbar(label='Z value')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2D Array Visualization')
plt.show()
```
