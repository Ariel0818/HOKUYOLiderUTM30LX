```python
import json
import numpy as np
```

```python
with open("segmentationlabel/1_1ColorImage.json", 'r') as file:
    data = json.load(file)
```

```python
total_area = 0

for shape in data['shapes']:
    points = np.array(shape['points'])
    # 计算多边形面积
    area = 0.5 * np.abs(np.dot(points[:, 0], np.roll(points[:, 1], 1)) - np.dot(points[:, 1], np.roll(points[:, 0], 1)))
    total_area += area
```

```python
print(total_area)
```
