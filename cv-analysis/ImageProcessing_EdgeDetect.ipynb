```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # 读取图片并转为灰度
image_grey = cv2.imread('DepthCamera_Data_Tanya/20 FEB 2023/1_1ColorImage.jpg', cv2.IMREAD_GRAYSCALE)
plt.axis('off')
plt.imshow(image_grey, cmap='gray') ## 
```

```python
# # image = cv2.imread('DepthCamera_Data_Tanya/20 FEB 2023/1_1ColorImage.jpg')  # BGR 读取通道
# image = cv2.imread('edgeDetectoutput/canny_51.5_103.0.jpg')  # BGR 读取通道
image = cv2.imread('edgeDetectoutput/canny_42.5_85.0.jpg')  # BGR 读取通道
print(image.shape)
b, g, r = cv2.split(image)
blue_image = image[:,:,0]  # blue 提取第一个通道
print(blue_image.shape)
plt.axis('off')
plt.imshow(b)
```

### canny Otsu's Binarization-- better

```python
high_thred, _ = cv2.threshold(blue_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
low_thred = 0.5 * high_thred
```

### canny median value

```python
median = np.median(b)
high_thred = median
low_thred = 0.5 * median
```

---

```python
# 使用Canny边缘检测， 需要在灰度图片上进行
edges = cv2.Canny(b, low_thred, high_thred)
```

```python
# 查找轮廓 需要在灰度图片上进行
contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# 在彩色图像上绘制轮廓
image = cv2.imread('DepthCamera_Data_Tanya/20 FEB 2023/1_1ColorImage.jpg') 
cv2.drawContours(image, contours, -1, (0, 0, 255), 2)  # 使用红色绘制轮廓

# 使用matplotlib显示图像
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# save
filename = f"edgeDetectoutput/canny_{low_thred}_{high_thred}.jpg"
cv2.imwrite(filename, image)
```

```python
areas = [cv2.contourArea(contour) for contour in contours]
```

```python
# 打印所有找到的轮廓的面积
for i, area in enumerate(areas):
    print(f"Area of contour {i}: {area} pixels")
```

## 去除背景的噪点

```python
out_image = cv2.imread('edgeDetectoutput/canny_42.5_85.0.jpg', cv2.IMREAD_GRAYSCALE)
_, binary = cv2.threshold(out_image, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(out_image)
binary.shape
```

```python
num_labels, labels = cv2.connectedComponents(binary)
```

```python
for label in range(1, num_labels):
    component = np.zeros(binary.shape, dtype=np.uint8)
    component[labels == label] = 255
    
    if cv2.countNonZero(component) < 6:
        binary[labels == label] = 0
```
